<h2>Outliers</h2>

```python
l = df.columns.values
number_of_columns= len(df.columns)
number_of_rows = len(l)-1//number_of_columns
print (number_of_columns)
print (number_of_rows)
df.columns[2]
```

<a href="https://seaborn.pydata.org/generated/seaborn.boxplot.html">Seaborn BoxPlot reference.</a>

```python
sns.boxplot(x=df["residual sugar"], notch=False, showcaps=False,
    flierprops={"marker": "x"},
    boxprops={"facecolor": (.4, .6, .8, .5)},
    medianprops={"color": "coral"},)
```

```python
plt.figure(figsize=(number_of_columns,5*number_of_rows))

for i in range(0,len(l)):
    plt.subplot(number_of_rows + 1,number_of_columns,i+1)
    sns.set_style('whitegrid')
    sns.boxplot(df[l[i]],orient='v')
    plt.tight_layout()
```

<h2>Label Encoding</h2>

Dataset: https://www.kaggle.com/datasets/manishkc06/engineering-graduate-salary-prediction

```python
engg_df = pd.read_csv("Engineering_graduate_salary.csv") 
engg_df.head()
```

```python
engg_df.Specialization.value_counts()
```

Categorial, Nominal values in columns usually can't be used with pre-built models. So, better to encode them. Scikit Learn provides a module for this.

Use pip to install sklearn - _pip install scikit-learn_

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
engg_df.Gender = le.fit_transform(engg_df.Gender)
engg_df
```

```python
le.fit(engg_df['10board'])
le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print(le_name_mapping)
```
